```python
import numpy as np
import matplotlib.pyplot as plt
import dolfin as dl

# https://github.com/NickAlger/helper_functions/blob/master/circle_mesh.py
from circle_mesh import circle_mesh
```

```python
mesh_center = np.array([-0.5, 0.5])
mesh_radius = np.sqrt(2) / 2.
mesh_h = 0.1

mesh = circle_mesh(mesh_center, mesh_radius, mesh_h)

dl.plot(mesh)
```

```python
# https://github.com/NickAlger/helper_functions/blob/master/pointcloud_nearest_neighbor.py
from pointcloud_nearest_neighbor import pointcloud_nearest_neighbor

pp = mesh.coordinates()
_, dd = pointcloud_nearest_neighbor(pp, return_min_distances=True)

min_h = np.min(dd)
median_h = np.median(dd)
max_h = np.max(dd)

print('min_h=', min_h, ', median_h=', median_h, ', max_h=', max_h)
```
