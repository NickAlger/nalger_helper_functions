```python
import numpy as np
import dolfin as dl
import matplotlib.pyplot as plt

# https://github.com/NickAlger/helper_functions/blob/master/box_mesh_nd.py
from box_mesh_nd import box_mesh_nd
```

```python
min1 = np.array([0.1])
max1 = np.array([0.9])
grid_shape1 = tuple([5])

mesh1 = box_mesh_nd(min1, max1, grid_shape1)
dl.plot(mesh1)
plt.title('1D box mesh')
```

```python
min2 = np.array([0.1, -0.1])
max2 = np.array([0.9, 1.1])
grid_shape2 = tuple([5, 7])

mesh2 = box_mesh_nd(min2, max2, grid_shape2)
dl.plot(mesh2)
plt.title('2D box mesh')
```

```python
min3 = np.array([0.0, 1.0, 2.0])
max3 = np.array([0.5, 1.5, 2.5])
grid_shape3 = tuple([3, 4, 5])

mesh3 = box_mesh_nd(min3, max3, grid_shape3)

plt.figure()

dl.plot(mesh3)
plt.title('3D box mesh')
```
