```python
import matplotlib.pyplot as plt

# https://github.com/NickAlger/helper_functions/blob/master/plot_ellipse.py
from plot_ellipse import plot_ellipse

# https://github.com/NickAlger/helper_functions/blob/master/plot_rectangle.py
from plot_rectangle import plot_rectangle


fig = plt.figure()

plot_ellipse(mu, Sigma, tau)
plot_rectangle(grid_min, grid_max)

plt.xlim(0.0, 0.8)
plt.ylim(0.0, 0.8)
plt.show()
```
