```python
import numpy as np
import matplotlib.pyplot as plt

# https://github.com/NickAlger/helper_functions/blob/master/make_regular_grid.py
from make_regular_grid import make_regular_grid
```

```python
min_pt = np.array([0.1, -0.1])
max_pt = np.array([1.1, 1.3])

grid_shape = (6,8)

(xx, yy), (X, Y) = make_regular_grid(min_pt, max_pt, grid_shape)

print('xx=', xx)
print('yy=', yy)

plt.figure()
plt.pcolor(X, Y, X)
plt.title('X')

plt.figure()
plt.pcolor(X, Y, Y)
plt.title('Y')

plt.figure()
plt.scatter(X.reshape(-1), Y.reshape(-1))

buf = 0.03
for k, txt in enumerate(np.arange(np.prod(grid_shape))):
    plt.gca().annotate(txt, (X.reshape(-1)[k]+buf, Y.reshape(-1)[k]))

plt.title('(X[:], Y[:])')
```
