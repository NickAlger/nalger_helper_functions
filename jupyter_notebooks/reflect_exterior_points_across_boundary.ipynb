```python
import numpy as np
import dolfin as dl
import matplotlib.pyplot as plt

from nalger_helper_functions import reflect_exterior_points_across_boundary, circle_mesh

mesh = circle_mesh(np.array([0.0, 0.0]), 1.0, 0.01)

square_mesh = dl.UnitSquareMesh(15, 16)
square_mesh.coordinates()[:] = 2.2 * square_mesh.coordinates() - 1.1
pp = square_mesh.coordinates()

plt.figure()
dl.plot(mesh)
dl.plot(square_mesh)
plt.plot(pp[:,0], pp[:,1], '.r')

pp_reflected = reflect_exterior_points_across_boundary(pp, mesh)

square_mesh.coordinates()[:] = pp_reflected

plt.figure()
# dl.plot(mesh)
dl.plot(square_mesh)
plt.plot(pp_reflected[:,0], pp_reflected[:,1], '.r')
```
