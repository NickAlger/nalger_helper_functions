```python
import numpy as np
import matplotlib.pyplot as plt
from nalger_helper_functions import *

mu = np.array([0.3, 0.5])
Sigma = np.array([[1.0, -0.1],[-0.1, 1.5]])
tau = 0.75
E_min, E_max = ellipsoid_bounding_box(mu, Sigma, tau)
E_shape = (18,23)
E = ellipsoid_characteristic_function(E_min, E_max, E_shape, mu, Sigma, tau)
E.plot()
plot_ellipse(mu, Sigma, tau)
plt.xlim([E_min[0]-0.1, E_max[0]+0.1])
plt.ylim([E_min[1]-0.1, E_max[1]+0.1])

inside_E = point_is_in_ellipsoid(E.gridpoints, mu, Sigma, tau)
outside_E = np.logical_not(inside_E)
plt.scatter(E.gridpoints[outside_E,0], E.gridpoints[outside_E,1], s=1, color='b')
plt.scatter(E.gridpoints[inside_E,0], E.gridpoints[inside_E,1], s=1, color='r')
```
