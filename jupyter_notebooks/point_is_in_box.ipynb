```python
import numpy as np

from nalger_helper_functions import point_is_in_box
```

```python
num_pts = 300

pp = np.random.randn(num_pts, 2)

box_min = np.array([-0.9, -0.7])
box_max = np.array([0.8, 1.1])

inside_inds = point_is_in_box(pp, box_min, box_max)
outside_inds = np.logical_not(inside_inds)

points_inside_box = pp[inside_inds, :]
points_outside_box = pp[outside_inds, :]
```

```python
import matplotlib.pyplot as plt

from nalger_helper_functions import plot_rectangle

plt.figure()
plt.scatter(points_inside_box[:,0], points_inside_box[:,1], c='r')
plt.scatter(points_outside_box[:,0], points_outside_box[:,1], c='b')
plot_rectangle(box_min, box_max)
```
