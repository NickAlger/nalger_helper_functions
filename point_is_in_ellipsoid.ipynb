```python
import numpy as np

# https://github.com/NickAlger/helper_functions/blob/master/point_is_in_ellipsoid.py
from point_is_in_ellipsoid import point_is_in_ellipsoid
```

```python
mu = np.array([0.41, 0.48])
Sigma = np.array([[2.1, 0.2],
                  [0.2, 1.6]])*0.1
tau = 2.3

num_pts = 200

pp = np.random.randn(num_pts, 2)

inside_inds = point_is_in_ellipsoid(pp, mu, Sigma, tau)
outside_inds = np.logical_not(inside_inds)

points_inside_ellipse = pp[inside_inds,:]
points_outside_ellipse = pp[outside_inds,:]
```

```python
import matplotlib.pyplot as plt

# https://github.com/NickAlger/helper_functions/blob/master/plot_ellipse.py
from plot_ellipse import plot_ellipse

plt.figure()

plt.scatter(points_inside_ellipse[:,0], points_inside_ellipse[:,1], c='r')
plt.scatter(points_outside_ellipse[:,0], points_outside_ellipse[:,1], c='b')

plot_ellipse(mu, Sigma, tau)
```
